```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer


```

```python
data = pd.read_csv(r'./housing.csv')

encoder = OneHotEncoder(sparse=False)
ocean_proximity_encoded = encoder.fit_transform(data[['ocean_proximity']])
data = data.drop('ocean_proximity', axis=1)
data = pd.concat([data, pd.DataFrame(ocean_proximity_encoded, columns=encoder.get_feature_names_out(['ocean_proximity']))], axis=1)

```

```python
X = data.drop("median_house_value", axis=1)
y = data["median_house_value"]
```

```python
missing_columns = X.columns[X.isnull().any()]
imputers = {}
for column in missing_columns:
    imputer = SimpleImputer(strategy='median')  # You can choose another strategy
    X[column] = imputer.fit_transform(X[[column]])
    imputers[column] = imputer

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
plt.figure(figsize=(15, 10))
plot_tree(clf, filled=True, feature_names=X.columns, class_names=True)
plt.show()
```

```python
y_pred = clf.predict(X_test)

correctly_classified = X_test[y_pred == y_test]
incorrectly_classified = X_test[y_pred != y_test]
num_correct = len(correctly_classified)
num_incorrect = len(incorrectly_classified)

print(f"Correctly Classified Instances: {num_correct}")
print(f"Incorrectly Classified Instances: {num_incorrect}")
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
naive_bayes_classifier = GaussianNB()
naive_bayes_classifier.fit(X_train, y_train)
y_pred_naive_bayes = naive_bayes_classifier.predict(X_test)
accuracy_naive_bayes = accuracy_score(y_test, y_pred_naive_bayes)
print(f"Naive Bayes Accuracy: {accuracy_naive_bayes:.2f}")
```

```python
y_pred = naive_bayes_classifier.predict(X_test)

correctly_classified = X_test[y_pred == y_test]
incorrectly_classified = X_test[y_pred != y_test]
num_correct = len(correctly_classified)
num_incorrect = len(incorrectly_classified)

print(f"Correctly Classified Instances: {num_correct}")
print(f"Incorrectly Classified Instances: {num_incorrect}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(X_train, y_train)
y_pred_knn = knn_classifier.predict(X_test)

accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f"KNN Accuracy: {accuracy_knn:.2f}")
```

```python
y_pred = knn_classifier.predict(X_test)

correctly_classified = X_test[y_pred == y_test]
incorrectly_classified = X_test[y_pred != y_test]
num_correct = len(correctly_classified)
num_incorrect = len(incorrectly_classified)

print(f"Correctly Classified Instances: {num_correct}")
print(f"Incorrectly Classified Instances: {num_incorrect}")
```
